### **B. Apakah ada kolom yang memiliki nilai kosong? Jika ada, apa saja?** <br>
Berdasarkan output fungsi info() diketahui bahwa semua variabel memiliki nilai missing value terkecuali variabel LoanNr_ChkDgt. Untuk mempermudah analisis akan dibuat fungsi untuk mengecek nilai missing value dengan source code sebagai berikut :

```python
def cek_null(df):
    col_na = df.isnull().sum().sort_values(ascending=False)
    percent = col_na / len(df)
    
    missing_data = pd.concat([col_na, percent], axis=1, keys=['Total', 'Percent'])
    if (missing_data[missing_data['Total'] > 0].shape[0] == 0):
        print("Tidak ditemukan missing value pada dataset")
        
    else:
        print(missing_data[missing_data['Total'] > 0])
    
cek_null(df)
```

Hasil fungsi cek_null(df) menghasilkan jumlah baris yang memiliki nilai missing value serta proporsinya terhaap dataset. Dapat dilihat bahwa terdapat variabel ChgOffDate yang memiliki nilai missing value > 80% dari dataset sehingga menggunakan variabel ini untuk analisis lebih lanjut tidak disarankan. Sedangkan untuk variabel lainnya memiliki missing value nya < 0,006% dari dataset, maka penanganan missing value ini bisa ditangani dengan drop/menghapus baris yang memiliki missing value atau inputasi nilai mean/median pada variabel yang bertipe data int/float.

Keempat variabel tersebut memiliki beberapa outlier sehingga perlu dilakukan penanganan outlier sebelum dilakukan pemodelan.


```python
# mengecek keberadaan outlier
df.boxplot(vert=False,figsize=(20,20))
```

```python
def outlier_checker(variable):
    all_outliers = []
    variable_to_check = df[variable]
    
    q1 = variable_to_check.quantile(0.25)
    q3 = variable_to_check.quantile(0.75)
    iqr = q3-q1

    lower_fence = q1-(1.5*iqr)
    upper_fence = q3+(1.5*iqr)

    sum_of_outlier = ((variable_to_check < lower_fence) | (variable_to_check > upper_fence)).sum()
    
    for row in variable_to_check:
        if((row < lower_fence) | (row > upper_fence)):
            all_outliers.append(row)
         
    percent = sum_of_outlier / len(variable_to_check)
    
    print(variable.upper()+":")
    print("Jumlah Outlier = ", sum_of_outlier)
    print("Outlier percentage = ", percent)
    print("List of outliers = ", all_outliers)

for col in df.select_dtypes(np.number).columns:
    outlier_checker(col)
    print()
```
